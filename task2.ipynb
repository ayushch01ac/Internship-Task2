To begin our program, we will first import pandas and read the downloaded excel file renamed to input 1 using read excel function to make a job easier by converting it into a dataframe

```python
import pandas as pd
raw_data = pd.read_excel(r"Input_1.xlsx")
```

Now we can see below that the data has been converted to a pandas dataframe 

```python
raw_data
```

Now we will take first three columns from the original data as they are same in the new format and store them in a new variable data.


```python
data = []
for index, row in raw_data.iterrows():
    data.append(row[['Name', 'id', 'Chapter Tag']])
```

```python
data
```

Now we will create three empty lists and we will store the values of 1st column, 2nd column and 3rd column in them respectively.

```python
names, username, chaptertag = [], [], []
```

Now since there are 9 tests for one student so firstly we will create 9 rows for each student with first 3 given columns using values stored in **data** and later we can delete the rows with empty or not applicable values if we come across them

```python
for val in data:
    for i in range(9):
        names.append(val[0])
        username.append(val[1])
        chaptertag.append(val[2])
```

Now, we will create a new pandas DataFrame named **df** according to output specifications and insert into it values column by column

```python
df = pd.DataFrame()
```

```python
df['Name'] = names
df['Username'] = username
df['Chapter Tag'] = chaptertag
```

```python
df
```

Now, we will create values for Test_Name column for 1 student and then multiply it by **length of (raw_data)** to get it for rest of the students too

```python
testnames = ['Concept Test 1','Concept Test 2','Concept Test 3','Concept Test 4','Concept Test 5',
                   'Full Chapter Test 1','Full Chapter Test 2','Topic Test 1','Topic Test 2'] * len(raw_data)
```

Now new column can be added in the dataframe using these values

```python
df['Test_Name'] = testnames
```

```python
df
```

Now we will create 6 empty lists for the remaining 6 columns and then we will iterate through each row and append values in those empty lists accordingly so that they can be used to create new columns in the dataframe.

```python
answered, correct, score, skipped, time_taken, wrong = [], [], [], [], [], []
```

All of this code could have been done in one single iteration rathen than 6 but for better code readability and understanding it has been separated.

```python
for index, row in raw_data.iterrows():
    answered.append(raw_data.iloc[index]['Concept Test 1 - answered'])
    answered.append(raw_data.iloc[index]['Concept Test 2 - answered'])
    answered.append(raw_data.iloc[index]['Concept Test 3 - answered'])
    answered.append(raw_data.iloc[index]['Concept Test 4 - answered'])
    answered.append(raw_data.iloc[index]['Concept Test 5 - answered'])
    answered.append(raw_data.iloc[index]['Full Chapter Test 1 - answered'])
    answered.append(raw_data.iloc[index]['Full Chapter Test 2 - answered'])
    answered.append(raw_data.iloc[index]['Topic Test 1 - answered'])
    answered.append(raw_data.iloc[index]['Topic Test 2 - answered'])
```

```python
for index, row in raw_data.iterrows():
    correct.append(raw_data.iloc[index]['Concept Test 1- correct'])
    correct.append(raw_data.iloc[index]['Concept Test 2- correct'])
    correct.append(raw_data.iloc[index]['Concept Test 3- correct'])
    correct.append(raw_data.iloc[index]['Concept Test 4- correct'])
    correct.append(raw_data.iloc[index]['Concept Test 5- correct'])
    correct.append(raw_data.iloc[index]['Full Chapter Test 1- correct'])
    correct.append(raw_data.iloc[index]['Full Chapter Test 2- correct'])
    correct.append(raw_data.iloc[index]['Topic Test 1- correct'])
    correct.append(raw_data.iloc[index]['Topic Test 2- correct'])
```

```python
for index, row in raw_data.iterrows():
    score.append(raw_data.iloc[index]['Concept Test 1 - score'])
    score.append(raw_data.iloc[index]['Concept Test 2 - score'])
    score.append(raw_data.iloc[index]['Concept Test 3 - score'])
    score.append(raw_data.iloc[index]['Concept Test 4 - score'])
    score.append(raw_data.iloc[index]['Concept Test 5 - score'])
    score.append(raw_data.iloc[index]['Full Chapter Test 1 - score'])
    score.append(raw_data.iloc[index]['Full Chapter Test 2 - score'])
    score.append(raw_data.iloc[index]['Topic Test 1 - score'])
    score.append(raw_data.iloc[index]['Topic Test 2 - score'])
```

```python
for index, row in raw_data.iterrows():
    skipped.append(raw_data.iloc[index]['Concept Test 1- skipped'])
    skipped.append(raw_data.iloc[index]['Concept Test 2- skipped'])
    skipped.append(raw_data.iloc[index]['Concept Test 3- skipped'])
    skipped.append(raw_data.iloc[index]['Concept Test 4- skipped'])
    skipped.append(raw_data.iloc[index]['Concept Test 5- skipped'])
    skipped.append(raw_data.iloc[index]['Full Chapter Test 1- skipped'])
    skipped.append(raw_data.iloc[index]['Full Chapter Test 2- skipped'])
    skipped.append(raw_data.iloc[index]['Topic Test 1- skipped'])
    skipped.append(raw_data.iloc[index]['Topic Test 2- skipped'])
```

```python
for index, row in raw_data.iterrows():
    time_taken.append(raw_data.iloc[index]['Concept Test 1 - time-taken (seconds)'])
    time_taken.append(raw_data.iloc[index]['Concept Test 2 - time-taken (seconds)'])
    time_taken.append(raw_data.iloc[index]['Concept Test 3 - time-taken (seconds)'])
    time_taken.append(raw_data.iloc[index]['Concept Test 4 - time-taken (seconds)'])
    time_taken.append(raw_data.iloc[index]['Concept Test 5 - time-taken (seconds)'])
    time_taken.append(raw_data.iloc[index]['Full Chapter Test 1 - time-taken (seconds)'])
    time_taken.append(raw_data.iloc[index]['Full Chapter Test 2 - time-taken (seconds)'])
    time_taken.append(raw_data.iloc[index]['Topic Test 1 - time-taken (seconds)'])
    time_taken.append(raw_data.iloc[index]['Topic Test 2 - time-taken (seconds)'])
```

```python
for index, row in raw_data.iterrows():
    wrong.append(raw_data.iloc[index]['Concept Test 1- wrong'])
    wrong.append(raw_data.iloc[index]['Concept Test 2- wrong'])
    wrong.append(raw_data.iloc[index]['Concept Test 3- wrong'])
    wrong.append(raw_data.iloc[index]['Concept Test 4- wrong'])
    wrong.append(raw_data.iloc[index]['Concept Test 5- wrong'])
    wrong.append(raw_data.iloc[index]['Full Chapter Test 1- wrong'])
    wrong.append(raw_data.iloc[index]['Full Chapter Test 2- wrong'])
    wrong.append(raw_data.iloc[index]['Topic Test 1- wrong'])
    wrong.append(raw_data.iloc[index]['Topic Test 2- wrong'])
```

Now we will make new columns using the above created lists

```python
df['answered'] = answered
df['correct'] = correct
df['score'] = score
df['skipped'] = skipped
df['time-taken (seconds)'] = time_taken
df['wrong'] = wrong
```

```python
df
```

Now we just have to remove the not applicable rows which don't have any value for tests so for that we can just check one column of that row which has '-' as a value. We will use answered column for that.

```python
temp = df[(df['answered'] == '-')].index
```

```python
df.drop(temp,inplace=True)
```

```python
df
```

Now we just have to remove the index and then we export the dataframe to an excel fine and our work is done

```python
df.reset_index(drop=True)
```

```python
df.to_excel('final_result.xlsx',index=False)
```
